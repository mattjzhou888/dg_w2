```python
import pandas as pd
import xlrd
import numpy as np
```

# Dataframe Creation

```python
cab_df = pd.read_csv("Cab_Data.csv")
cab_df
```

```python
city_df = pd.read_csv("City.csv")
city_df
```

```python
customer_df = pd.read_csv("Customer_ID.csv")
customer_df
```

```python
transaction_df = pd.read_csv("Transaction_ID.csv")
transaction_df
```

```python
print(f"{cab_df.isna().to_numpy().sum()} {city_df.isna().to_numpy().sum()} {customer_df.isna().to_numpy().sum()} {transaction_df.isna().to_numpy().sum()}")
print(f"{cab_df.isnull().to_numpy().sum()} {city_df.isnull().to_numpy().sum()} {customer_df.isnull().to_numpy().sum()} {transaction_df.isnull().to_numpy().sum()}")
```

```python
df = pd.merge(cab_df, transaction_df, on="Transaction ID")
df = pd.merge(df, customer_df, on="Customer ID")
df
```

```python
dates = np.empty((4, len(df)), dtype=int)
for i in range(len(df)):
    date = xlrd.xldate_as_datetime(df.loc[i, "Date of Travel"], 0)
    dates[0, i] = date.year
    dates[1, i]= date.month
    dates[2, i] = date.day
    dates[3, i] = date.weekday()
df["Year"] = dates[0]
df["Month"] = dates[1]
df["Day"] = dates[2]
df["Weekday"] = dates[3]
df = df.sort_values(by=["Year", 'Month', "Day"])
df["Profit"] = df["Price Charged"] - df["Cost of Trip"]
df["Profit/KM"] = df["Profit"] / df["KM Travelled"]
df
```

```python
df_y = df.loc[df["Company"] == "Yellow Cab"]
df_p = df.loc[df["Company"] == "Pink Cab"]
print(f"# Yellow Cab Rides: {len(df_y)}, # Pink Cab Rides: {len(df_p)}")
```

# Analysis

```python
profit_y = df_y["Profit"].sum()
profit_p = df_p["Profit"].sum()
profit_km_y = df_y["Profit/KM"].mean()
profit_km_p = df_p["Profit/KM"].mean()
print(f"Yellow Total Profit: {profit_y}, Pink Total Profit: {profit_p}")
print(f"Yellow Profit/Ride: {profit_y / len(df_y)}, Pink Profit/Ride: {profit_p / len(df_p)}")
print(f"Yellow Profit/KM: {profit_km_y}, Pink Profit/KM: {profit_km_p}")
```

```python
for i in range(2016, 2019):
    yearly_profit_y = df_y[df_y["Year"] == i]["Profit"].sum()
    yearly_profit_p = df_p[df_p["Year"] == i]["Profit"].sum()
    yearly_num_rides_y = (df_y["Year"] == i).sum()
    yearly_num_rides_p = (df_p["Year"] == i).sum()
    yearly_profit_km_y = df_y[df_y["Year"] == i]["Profit/KM"].mean()
    yearly_profit_km_p = df_p[df_p["Year"] == i]["Profit/KM"].mean()
    print(f"Year {i}: Yellow Profit: {yearly_profit_y}, Pink Profit {yearly_profit_p}. Yellow Profit/Ride: {yearly_profit_y/yearly_num_rides_y}, Pink Profit/Ride: {yearly_profit_p/yearly_num_rides_p}")
    print(f"Year {i}: Yellow Profit/KM: {yearly_profit_km_y}, Pink Profit/KM {yearly_profit_km_p}")
```

```python
g10_y = (df_y["Customer ID"].value_counts() >= 10).sum()
g20_y = (df_y["Customer ID"].value_counts() >= 20).sum()
g10_p = (df_p["Customer ID"].value_counts() >= 10).sum()
g20_p = (df_p["Customer ID"].value_counts() >= 20).sum()
customers_y = df_y["Customer ID"].unique()
customers_p = df_p["Customer ID"].unique()
customer_df_y = customer_df[customer_df["Customer ID"].isin(customers_y)]
customer_df_p = customer_df[customer_df["Customer ID"].isin(customers_p)]
print(f"Yellow customers: {len(customers_y)}, Pink customers: {len(customers_p)}")
print(f"Yellow: g10 - {g10_y} ({g10_y/len(customers_y)}), g20 - {g20_y} ({g20_y/len(customers_y)}); \
    Pink: g10 - {g10_p} ({g10_p/len(customers_p)}), g20 - {g20_p} ({g20_p/len(customers_p)})")
```
